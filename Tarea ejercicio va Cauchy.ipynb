```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Definir la función de densidad de la distribución Cauchy estándar
f = lambda x: 1 / (np.pi * (1 + x**2))

# Fórmula como string
formula_f = "f(x) = 1 / (π (1 + x²))"
print("Fórmula de la función de densidad:", formula_f)

# Verificar que la integral sea 1
integral, error = quad(f, -np.inf, np.inf)
print(f"Integral de f(x) en (-∞, ∞): {integral:.4f} (error estimado: {error:.1e})")

# Resultado esperado: 1
```

```python
# Definir la función de distribución acumulada (CDF)
F = lambda x: (1/np.pi) * np.arctan(x) + 0.5

# Fórmula como string
formula_F = "F(x) = (1/π) arctan(x) + 0.5"
print("Fórmula de la función de distribución acumulada:", formula_F)
```

```python
# Crear un rango de valores de x
x = np.linspace(-10, 10, 1000)

# Graficar f(x)
plt.figure(figsize=(12, 5))

# Densidad
plt.subplot(1, 2, 1)
plt.plot(x, f(x), label=r'$f(x) = \frac{1}{\pi(1+x^2)}$', color='blue')
plt.title('Función de Densidad de Cauchy (param=1)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()

# CDF
plt.subplot(1, 2, 2)
plt.plot(x, F(x), label=r'$F(x) = \frac{1}{\pi} \arctan(x) + \frac{1}{2}$', color='green')
plt.title('Función de Distribución Acumulada de Cauchy')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Fórmula como string
formula_inversion = "x = tan(π (u - 0.5))"
print("Fórmula de generación aleatoria (inversión):", formula_inversion)

# Simular N variables aleatorias Cauchy
N = 10000
u = np.random.uniform(0, 1, N)
x_cauchy = np.tan(np.pi * (u - 0.5))

# Graficar histograma de los datos simulados
plt.hist(x_cauchy, bins=100, density=True, color='skyblue', alpha=0.7, label='Simulación')
plt.plot(x, f(x), 'r', lw=2, label='Densidad teórica')
plt.title('Simulación vs Densidad teórica (Cauchy)')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.legend()
plt.grid(True)
plt.xlim(-10, 10)
plt.show()
```
